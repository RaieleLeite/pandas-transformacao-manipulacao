## Dados de vendas

```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_json('/content/dados_vendas_clientes.json')
```

```python
df = pd.json_normalize(df['dados_vendas'])
df.head()
```

```python
colunas_df = list(df.columns)
```

```python
colunas_alterar = ['Cliente', 'Valor da compra']
df = df.explode(colunas_alterar[:])
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df.info()
```

```python
df['Valor da compra'] =  df['Valor da compra'].apply(lambda x : x.replace('R$', '').replace(',', '.').strip())
df['Valor da compra'] =  df['Valor da compra'].astype(np.float64)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Cliente'] = df['Cliente'].str.lower()
df['Cliente'] = df['Cliente'].str.replace('[^a-z ]', '', regex=True).str.strip()
df.head()
```

```python
df['Data de venda'] = pd.to_datetime(df['Data de venda'], format='%d/%m/%Y')
```

```python
maiores_compradores = df.groupby(['Cliente'])['Valor da compra'].sum().sort_values(ascending=False)
maiores_compradores
```

## Dados Locação

```python
import pandas as pd
import numpy as np
import datetime
```

```python
df_locacao = pd.read_json('/content/dados_locacao_imoveis.json')
```

```python
df_locacao =  pd.json_normalize(df_locacao['dados_locacao'])
```

```python
colunas_alterar_locacao = ['datas_combinadas_pagamento', 'datas_de_pagamento','valor_aluguel' ]
```

```python
df_locacao = df_locacao.explode(colunas_alterar_locacao[:])
```

```python
df_locacao.reset_index(inplace=True, drop=True)
```

```python
df_locacao['valor_aluguel'] = df_locacao['valor_aluguel'].apply(lambda x: x.replace('$','').replace(',','.').replace('reais','').strip())
```

```python
df_locacao['valor_aluguel'] = df_locacao['valor_aluguel'].astype(np.float64)
```

```python
df_locacao.info()
```

```python
df_locacao.head()
```

```python
df_locacao['apartamento'] = df_locacao['apartamento'].str.replace(' \(blocoAP\)', '', regex=True)
df_locacao.head()
```

```python
df_locacao['datas_combinadas_pagamento'] = pd.to_datetime(df_locacao['datas_combinadas_pagamento'], format="%d/%m/%Y")
df_locacao['datas_de_pagamento'] = pd.to_datetime(df_locacao['datas_de_pagamento'], format="%d/%m/%Y")
df_locacao.info()
```

```python
df_locacao['dias_em_atraso'] = (df_locacao['datas_de_pagamento'] - df_locacao['datas_combinadas_pagamento']).dt.days.astype(np.int64)
```

```python
df_locacao['atrasou'] = df_locacao['dias_em_atraso'].apply(lambda x: True if x > 0 else False)
```

```python
lista_atrasados = df_locacao.groupby('apartamento')['atrasou'].sum()
lista_atrasados
```

```python
media_atraso = df_locacao.groupby(['apartamento'])['dias_em_atraso'].mean()
media_atraso
```
